```python
# 練習1:基礎定義
tool_name = "Python"
version = 3.12
is_useful = True

# 使用f-string(現代 Python最推薦的寫法) 來組合字串
print(f"我正在學習 {tool_name}，目前的版本是 {version}。這很有用嗎? {is_useful}")

#練習2:建立學習清單
skills = ["Git", "Colab", "Python", "Linux"]

#使用for迴圈逐一印出
for skill in skills:
  if skill == "Python":
    print(f"★ 發現重點項目：{skill}")
  else:
    print(f"已掌握技能：{skill}")

#練習3:模擬使用者資料
user_info = {
    "name": "Chao-hsun",
    "role": "Engineer",
    "level": 1
}
#取得特定數值
print(f"工程師名稱:{user_info['name']}")

#新增一個欄位
user_info["status"] = "Learning"
print(user_info)

%cd /content/drive/MyDrive/python-learning
!git add .
!git commit -m "完成第一課：基礎語法練習"
!git push origin main
```
