```python
# 練習1:基礎定義
tool_name = "Python"
version = 3.12
is_useful = True

# 使用f-string(現代 Python最推薦的寫法) 來組合字串
print(f"我正在學習 {tool_name}，目前的版本是 {version}。這很有用嗎? {is_useful}")

#練習2:建立學習清單
skills = ["Git", "Colab", "Python", "Linux"]

#使用for迴圈逐一印出
for skill in skills:
  if skill == "Python":
    print(f"★ 發現重點項目：{skill}")
  else:
    print(f"已掌握技能：{skill}")

#練習3:模擬使用者資料
user_info = {
    "name": "Chao-hsun",
    "role": "Engineer",
    "level": 1
}
#取得特定數值
print(f"工程師名稱:{user_info['name']}")

#新增一個欄位
user_info["status"] = "Learning"
print(user_info)

%cd /content/drive/MyDrive/python-learning
!git add .
!git commit -m "完成第一課：基礎語法練習"
!git push origin main
```

```python
# 練習：建立一個伺服器清單
servers = [
    {"hostname": "WEB-01", "ip": "192.168.1.10", "status": "Online"},
    {"hostname": "DB-01", "ip": "192.168.1.20", "status": "Offline"},
    {"hostname": "AD-01", "ip": "192.168.1.30", "status": "Online"}
]

print("--- 異常設備檢查報告 ---")
# 練習：使用迴圈找出 Offline 的設備
for s in servers:
    if s["status"] == "Offline":
        print(f"警報：伺服器 {s['hostname']} ({s['ip']}) 目前狀態為： {s['status']}！")
```

```python
# 定義一個檢查權限的函式
def check_permission(user_role):
    if user_role == "Admin":
        return "准許進入系統管理頁面"
    else:
        return "權限不足，僅限管理者存取"

# 測試函式
current_user = "Engineer"
result = check_permission(current_user)
print(f"檢查結果：{result}")
```
